```python
data = [[1, 'Alice'], [2, 'Bob'], [13, 'John'], [6, 'Alex']]
students = pd.DataFrame(data, columns=['student_id', 'student_name']).astype({'student_id':'Int64', 'student_name':'object'})
data = [['Math'], ['Physics'], ['Programming']]
subjects = pd.DataFrame(data, columns=['subject_name']).astype({'subject_name':'object'})
data = [[1, 'Math'], [1, 'Physics'], [1, 'Programming'], [2, 'Programming'], [1, 'Physics'], [1, 'Math'], [13, 'Math'], [13, 'Programming'], [13, 'Physics'], [2, 'Math'], [1, 'Math']]
examinations = pd.DataFrame(data, columns=['student_id', 'subject_name']).astype({'student_id':'Int64', 'subject_name':'object'})
```

```python
df = pd.merge(students, subjects, how='cross')
df_num = examinations.groupby(['student_id', 'subject_name']).size().reset_index(name='attended_exams')
df_final = pd.merge(df, df_num, on=['student_id', 's'], how='left').fillna(0)
df_final['attended_exams'] = df_final['attended_exams'].astype(int)
df_final
```

```python
students['key'] = 1
subjects['key'] = 1
all_combinations = pd.merge(students, subjects, on='key', how='inner').drop(columns='key')

exams_count = examinations.groupby(['student_id', 'subject_name']).size().reset_index(name='attended_exams')
final_df = pd.merge(all_combinations, exams_count, on=['student_id', 'subject_name'], how='left')
final_df['attended_exams'] = final_df['attended_exams'].fillna(0).astype(int)
final_df = final_df.sort_values(by=['student_id', 'subject_name'])
final_df
```

```python
data = [[101, 'John', 'A', None], [102, 'Dan', 'A', 101], [103, 'James', 'A', 101], [104, 'Amy', 'A', 101], [105, 'Anne', 'A', 101], [106, 'Ron', 'B', 101]]
employee = pd.DataFrame(data, columns=['id', 'name', 'department', 'managerId']).astype({'id':'Int64', 'name':'object', 'department':'object', 'managerId':'Int64'})
df = employee.groupby('managerId').size().reset_index(name='cnt')
df2 = df[df['cnt']>=5]['managerId']    
df3 = employee[employee['id'].isin(list(df2))][['name']]
df2.index
```

```python
subordinate_count = employee.groupby('managerId')['id'].count()

# 找出直接下属数量大于等于5的经理
managers_with_5_subordinates = subordinate_count[subordinate_count >= 5].index

# 使用这些经理的id来获取经理姓名
result = employee[employee['id'].isin(managers_with_5_subordinates)]['name']

# 将 Series 转换为 DataFrame
result_df = result.to_frame(name='name')
```
